```python
### 8. Hands-On Part 1: The PDF Explorer

#Time to put it all together! Let's write a reusable function that loads  
#any PDF and gives us a nice summary. This is something you'll actually  
#use in real projects.
from langchain_community.document_loaders import PyPDFLoader


def explore_pdf(file_path):
    """load pdf document for nice summary"""
    pdf_path = PyPDFLoader(file_path = file_path)
    pdf_documents = pdf_path.load()
    
#summer of pdf 
    summary = f"PDF Summary:\nTotal Pages: {len(pdf_documents)}\n\n"
    for i, doc in enumerate(pdf_documents):
        summary += f"Page {i+1}:\nContent Preview: {doc.page_content[:200]}...\nMetadata: {doc.metadata}\n\n"
    return summary

   # min(3, len(documents)) picks the SMALLER value
    # If the PDF has 50 pages, we show 3. If it has 2 pages, we show 2.
    # This prevents crashing when the PDF has fewer than 3 pages.
    pages_to_show = min(3, len(pdf_documents))
    
    for i in range(pages_to_show):
        doc = pdf_documents[i]
        summary += f"Page {i+1}:\nContent Preview: {doc.page_content[:200]}...\nMetadata: {doc.metadata}\n\n"
    return summary
    # show what metadata key are available 
    if pdf_documents:
        available_metadata_keys = set()
        for doc in pdf_documents:
            available_metadata_keys.update(doc.metadata.keys())
        summary += f"Available Metadata Keys: {available_metadata_keys}\n"
    else:
        summary += "No pages found in the PDF.\n"
    
    print(summary)
# test the function with sample pdf
pdf_summary = explore_pdf("data/who_substances_surveillance.pdf")
print(pdf_summary)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://radiopaedia.org/search?scope=cases&sort=date_of_publication"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

links = []
for a in soup.find_all("a", href=True):
    if "/cases/" in a["href"]:
        full_link = "https://radiopaedia.org" + a["href"]
        links.append(full_link)

links = list(set(links))  # remove duplicates
print(links[:5])
```

```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader(links[:5])  # test with few links
docs = loader.load()
```

```python
from bs4 import BeautifulSoup

def extract_findings(html):
    soup = BeautifulSoup(html, "html.parser")
    
    # try common patterns
    findings = soup.find("div", string=lambda x: x and "Findings" in x)
    
    return findings.text if findings else "No findings found"

# call the function on the first document's content
findings_text = extract_findings(url)
print(findings_text)
```
